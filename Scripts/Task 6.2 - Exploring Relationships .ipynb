```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
df_cacao_cleaned = pd.read_csv('../Data/Updated/df_cacao_cleaned.csv')
```

2. Exploratory Visualizations

    3. Correlation Heatmap

```python
df_cacao_cleaned_numeric = df_cacao_cleaned.select_dtypes(include=[float, int])
```

```python
correlation_matrix = df_cacao_cleaned_numeric.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

4. Scatterplots

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_cacao_cleaned_numeric['Rating'], df_cacao_cleaned_numeric['Reference_Number'], alpha=0.5)
plt.title('Scatterplot of Rating vs Reference_Number')
plt.xlabel('Rating')
plt.ylabel('Reference_Number')
plt.show()
```

No clear trend, just shows that ratings are given consistently with the reference numbers

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_cacao_cleaned_numeric['Rating'], df_cacao_cleaned_numeric['Review_Date'], alpha=0.5)
plt.title('Scatterplot of Rating vs Review_Date')
plt.xlabel('Rating')
plt.ylabel('Review_Date')
plt.show()
```

No strong correlation but shows consistent ratings

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_cacao_cleaned_numeric['Review_Date'], df_cacao_cleaned_numeric['Reference_Number'], alpha=0.5)
plt.title('Scatterplot of Review_Date vs Reference_Number')
plt.xlabel('Review_Date')
plt.ylabel('Reference_Number')
plt.show()
```

strong relationship as reference number goes up so does the review_date - shows that the higher the reference number the more recent the review
